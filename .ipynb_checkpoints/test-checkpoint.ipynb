```python
# -*- coding: utf-8 -*-

import pandas as pd
from shapely.geometry import Point, shape

from flask import Flask
from flask import render_template
import json

import geopandas as gpd
from shapely.ops import cascaded_union
import pickle
```

```python
census = gpd.read_file('./input/census.geojson')
```

```python
all_df = pd.read_csv('./input/cencode_index.csv')
```

```python
census['cen_code'] = census['GEOID'].astype(int)
gdf = census.merge(all_df, left_on='cen_code', right_on='cen_code')
len(gdf)
```

```python
gdf['cen_code']
```

```python

epsg_utm = 26910 #utm zone 10n
gdf = gdf.to_crs(epsg=epsg_utm)
wa = gpd.read_file('./input/neighborhoods.geojson').to_crs(epsg=epsg_utm)
polygons = wa['geometry'].values
wash = gpd.GeoSeries(cascaded_union(polygons)).iloc[0]
gdf['geometry'] = gdf.intersection(wash)
```
