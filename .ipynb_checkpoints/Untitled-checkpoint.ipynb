```python
import pandas
import numpy
from pygam import PoissonGAM, s, f
import matplotlib.pyplot as plt
```

```python
df = pandas.read_csv('./data/filt_mat.csv')
```

```python
catenc = pandas.factorize(df['zone'])
df['zone'] = catenc[0]
```

```python
y = df['cafe']
X = df[['daytime_density', 'daytime_pop', 'medinc', 'hs', 'phd', 'oldpop', 'youngpop', 'brewery', 'zone']]
```

```python
gam = PoissonGAM(s(0) + s(1) + s(2) + s(3) + s(4) + s(5) + s(6) + s(7) + s(8) + f(9)).fit(X,y)
```

```python
gam.summary()
```

```python
XX = gam.generate_X_grid(term=0, n=500)
```

```python

plt.plot(XX, gam.predict(XX), 'r--')
plt.title('95% prediction interval');
```

```python
for i, term in enumerate(gam.terms):
    if term.isintercept:
        continue

    XX = gam.generate_X_grid(term=i)
    pdep, confi = gam.partial_dependence(term=i, X=XX, width=0.95)

    plt.figure()
    plt.plot(XX[:, term.feature], pdep)
    plt.plot(XX[:, term.feature], confi, c='r', ls='--')
    plt.title(repr(term))
    plt.show()
```

```python
df['pred'] = gam.predict(X)
```

```python
df.plot(kind='scatter', x='cafe', y='pred');
```
